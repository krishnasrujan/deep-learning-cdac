```python
def print_pyramid(iter):
    for n in range(1,iter):
        #print("Iteration no: %r" % ( n ) )
        for x in range(1,n):
            #print("Inner loop: %r" % ( x ) )
            print ('*'*x)
        print("------")

print_pyramid(6)
```

Yes! to access *'iCount'* we need an instance of the Class Object... Same is true for function *'func'* as well

The instantiation of a class creates an empty object. Its is advisable to create objects with instances initialized with attributes. Hence, a class may define a special method named **__init__()**.

```python
class BetterClass:
    
    def __init__(self):
        self.icount = 12345
        self.msg = 'You have reached the function'
        
    def func(self, msg):
        return msg
    
    def __str__(self):
        return('{} | {}'.format(self.icount, self.msg))
```

```python
# Instantiate the class
b = BetterClass()

print(b)
```

```python
# Instantiate the class
b = BetterClass()

print ("This is icount:", b.icount)
```

```python
b.icount = 6789

print ("Revised icount is:", b.icount)
```

```python
print ("and we can access the method:", b.func("How are you?"))
```

```python
class AddClass:
    def __init__(self, x=1.0, y=2.0):
        self.one = x
        self.two = y
    def add(self):
        return (self.one+self.two)
```

```python
a = AddClass()
print ("Let's look at the values....")
print("Class initialized with values:" , a.one, a.two)
print("Sum of which is :", a.add())
```

```python
# Let's pass some values
b = AddClass(3.0, 4.5)

print ('with values passed at the initialization....')

print("You entered:", b.one, b.two)
print("Sum is :", b.add())
```

# Differential Equations

## Why??

We need to find minimum and maximum of some functions later in the course. We will be needing both differential equation as well as roots.

First principle of differential equations is simply

For any function f(x):

<div style="font-family: Arial; font-size:1.1em;color:red;">
$$
\begin{aligned}
f'(x) = \frac{d f(x)}{dx} = \lim_{\delta{x} \to 0} \frac{f(x +\delta{x}) - f(x)}{\delta{x}}\\
\end{aligned}
$$
</div>

Lets take  simple example. Our favourite sine curve.

```python
# lets create lot data points
N = 100
x  = np.linspace(0, 4*np.pi, N)
```

```python
# Really dont need to but for completeness, lets define a function to 
def fn_sine(x):
    return np.sin(x)
```

```python
# Calculate Y
y = fn_sine(x)
```

```python
# quick plot to ensure all is fine
```

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
dy = []
dx = 4 * np.pi / N 

# Lets calculate d(y)
for i in range( len(x)-1):
    tmp = (y[i+1] -y[i])/dx
    dy.append(tmp)
```

```python
len(x), len(y), len(dy)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y);
ax.plot(x[:-1], dy); # remember differentiation will have one point less
```

## Using Numpy libraries

### Numpy. gradient

```python
ndy = np.gradient(y, dx)
```

```python
fig, [ax1,ax2] = plt.subplots(1,2, figsize = (12,4))
ax1.plot(x,y);
ax1.plot(x[:-1], dy); # remember differentiation will have one point less
ax1.set_title('First Principle')
ax2.plot(x,y)
ax2.plot(x, ndy);
ax2.set_title('Numpy Gradient')
ax1.grid()
ax2.grid()
```

```python
# Lets define a function to 
def fn_sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-5, 5, N)
y = fn_sigmoid(x)
```

```python
dy = []
# Lets calculate d(y)
for i in range( len(x)-1):
    tmp = (y[i+1] -y[i])/dx
    dy.append(tmp)
```

```python
ndy = np.gradient(y, dx)
```

```python
fig, [ax1,ax2] = plt.subplots(1,2, figsize = (12,4))
ax1.plot(x,y);
ax1.plot(x[:-1], dy); # remember differentiation will have one point less
ax1.set_title('First Principle')
ax2.plot(x,y)
ax2.plot(x, ndy);
ax2.set_title('Numpy Gradient')
ax1.grid()
ax2.grid()
```

<div style="font-family: Arial; font-size:1.1em;color:black;">
$
\begin{aligned}
y = f(g(x))\\
\frac{dy}{dx}= \frac{df}{dg}\frac{dg}{dx}\\
Given:\\
y = \frac{1}{(1+e^{-x})}\\
y = \frac{1}{g}\\
g = {(1+e^{-x})}\\
\frac{dg}{dx} = -e^{-x}\\
\frac{dy}{dg} = -\frac{1}{g^2}\\
Hence:
\frac{dy}{dx}= \frac{df}{dg}\frac{dg}{dx}\\
\frac{dy}{dx}= (-e^{-x}).(-\frac{1}{(1+e^{-x})^2})\\
= \frac{e^{-x}}{(1+e^{-x})^2}\\
\end{aligned}
$
