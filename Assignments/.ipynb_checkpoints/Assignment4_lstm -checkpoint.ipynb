```
# There are ?'s in some columns which represent null values. So we will read the data again by replacing ? with null 
```

```
data = pd.read_csv('/content/drive/MyDrive/household_power_consumption.txt',sep=';',na_values='?',parse_dates=[['Date','Time']])
data.head()
```

```
data.info()
```

```
# There are null values in sub_metering_3 column
data.isnull().any()
```

```
# Since this is a time series data we will fill the null values with forward fill method

data = data.ffill()

# Now there won't be any null values
data.isnull().any()
```

```
data.set_index('Date_Time',inplace=True)
data.head()
```

```
data = data.resample('H').sum()
data.head()
```

```
# Let us see the trend and seasonality in the data for our target feature

plt.figure(figsize=(20,8))
plt.plot(data['Global_active_power'])
```

```
data['target'] = data['Global_active_power'].shift(-1)
data.head()
```

```
data['Sub_metering_1'].value_counts()
```

```
# last row has a null value in target feature so we will drop that row
data.dropna(axis=0,inplace=True)
data.isnull().any()
```

```
# considering 80% train data and 20% test data

train_size = int(data.shape[0]*0.8)
X = data.drop('target',axis=1)
y = data['target']

train_x = X.iloc[:train_size,:]
test_x = X.iloc[train_size:,:]
train_y = y.iloc[:train_size]
test_y = y.iloc[train_size:]
```

```
from sklearn.preprocessing import MinMaxScaler

def scaler(X,y):

  # defining separate objects for input and target features. It will help in inverse transforming while plotting the predictions
  input = MinMaxScaler()  
  target = MinMaxScaler()
  X = input.fit_transform(X)
  y = target.fit_transform(np.array(y).reshape(-1,1))  
  return X,y,target  # returning target scaling obj for inverse scaling the predictions

train_x,train_y,target_train_obj = scaler(train_x,train_y)
test_x,test_y,target_test_obj = scaler(test_x,test_y)
```

```
# reshaping the input dataset according the lstm's requirement

# since we are predicting output based on one time step,time_steps=1

time_steps = 1

train_x = train_x.reshape(train_x.shape[0],time_steps,train_x.shape[1])
test_x = test_x.reshape(test_x.shape[0],time_steps,test_x.shape[1])
```

## Model Building



```
import tensorflow as tf
from tensorflow.keras.layers import LSTM,Dropout,Dense
from tensorflow.keras import Sequential
```

```
def build_model(lr):
  model = Sequential()
  model.add(LSTM(50,input_shape=(time_steps,train_x.shape[2])))
  model.add(Dropout(0.2))
  model.add(Dense(1))
  adm = tf.keras.optimizers.Adam(lr)
  model.compile(loss='mse',optimizer=adm)
  return model
```

```
for lr in [0.01,0.05,0.001,0.0001]:
  model = build_model(lr)
  history = model.fit(train_x,train_y,epochs=15,verbose=0,validation_data=(test_x,test_y))
  plt.plot(history.history['loss'],c='b')
  plt.plot(history.history['val_loss'],c='r')
  plt.grid()
  plt.title(lr)
  plt.show()
```

```
# 0.001 seems to be the optimal learning rate 
model = build_model(0.001)
model.fit(train_x,train_y,epochs=15,validation_data=(test_x,test_y))
```

```
model.save('/content/drive/MyDrive/house_hold_lstm.h5')
```

```
lstm_model = tf.keras.models.load_model('/content/drive/MyDrive/house_hold_lstm.h5')
```

```
lstm_model.summary()
```

```
lstm_model.evaluate(test_x,test_y)
```

```
y_test_preds = lstm_model.predict(test_x)
```

```
y_test_preds = target_test_obj.inverse_transform(y_test_preds)
y_test = target_test_obj.inverse_transform(test_y)
```

```
plt.figure(figsize=(20,8))
plt.plot(y_test_preds)
plt.scatter(x = range(len(test_y)), y = y_test,c='g')
plt.ylabel('Global Active Power')
plt.show()
```

```
model = Sequential()
model.add(LSTM(50,input_shape=(time_steps,train_x.shape[2]),kernel_initializer='HeNormal'))
model.add(Dropout(0.2))
model.add(Dense(1))
adm = tf.keras.optimizers.Adam(0.001)
model.compile(loss='mse',optimizer=adm)
```

```
model.fit(train_x,train_y,epochs=15,validation_data=(test_x,test_y))
```

```
y_test_preds = model.predict(test_x)
```
